```python
from transformers import AutoTokenizer, AutoModel
import torch
from sklearn.metrics.pairwise import cosine_similarity

# Load SAPBERT
model_name = "cambridgeltl/SapBERT-from-PubMedBERT-fulltext"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
model.eval()
```

```python
def embed_text(texts):
    with torch.no_grad():
        inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")
        outputs = model(**inputs)
        embeddings = outputs.last_hidden_state[:, 0, :]  # CLS token
        return embeddings.cpu().numpy()
```

```python
import numpy as np
from tqdm import tqdm

# Simulate a larger KG concept list (realistic names)
kg_concepts = [
    "aspirin", "metformin", "heart failure", "chronic kidney disease", "diabetes",
    "stroke", "pneumonia", "seizures", "acetaminophen", "blood glucose",
    "renal failure", "infection", "hypertension", "hyperlipidemia", "anemia",
    "cardiac arrest", "COPD", "hepatitis", "dementia", "cancer"
]

# Re-embed the full KG concept list
kg_vecs = embed_text(kg_concepts)
kg_vecs /= np.linalg.norm(kg_vecs, axis=1, keepdims=True)

# Prepare attributes
attribute_names = [f"attribute_{i}" for i in range(max_attr_id + 1)]

# Container for matches
id_to_kg_match = {}

# Batch size
batch_size = 64

# Embed and link in batches
for i in tqdm(range(0, len(attribute_names), batch_size), desc="Linking attributes"):
    batch_attr_names = attribute_names[i:i+batch_size]
    attr_vecs = embed_text(batch_attr_names)
    attr_vecs /= np.linalg.norm(attr_vecs, axis=1, keepdims=True)

    sims = cosine_similarity(attr_vecs, kg_vecs)
    top_matches = sims.argmax(axis=1)

    for j, attr in enumerate(batch_attr_names):
        id_to_kg_match[i + j] = kg_concepts[top_matches[j]]
```

 Build Hypergraph Structure for MIMIC-III

```python
# Load visit → attribute_id hyperedges
# (already loaded previously)
# `hyperedges = [...]`

# Load attribute → KG entity mapping
import json
with open("attribute_to_kg_entity.json", "r") as f:
    id_to_kg_match = json.load(f)
```

```python
# Convert attribute IDs to KG entities for each visit
hypergraph_kg = []

for visit in hyperedges:
    kg_nodes = [id_to_kg_match[str(attr)] for attr in visit if str(attr) in id_to_kg_match]
    kg_nodes = list(set(kg_nodes))  # remove duplicates
    hypergraph_kg.append(kg_nodes)

# Check one example
print("Example hyperedge (KG entities):", hypergraph_kg[0])
```

✅ Step 7: Build Input Matrix for the Hypergraph Transformer


```python
import torch

embedding_dim = 50
max_nodes = max(len(h) for h in hypergraph_kg)

X = torch.zeros((len(hypergraph_kg), max_nodes, embedding_dim))
M = torch.zeros((len(hypergraph_kg), max_nodes))

for i, hyperedge in enumerate(hypergraph_kg):
    for j, node in enumerate(hyperedge):
        if node in entity_embeddings:
            X[i, j] = torch.tensor(entity_embeddings[node])
            M[i, j] = 1
```

```python
# Already loaded earlier
# `edge_labels = np.loadtxt(...).astype(int)`
labels = torch.tensor(edge_labels)  # shape: (n_visits, n_classes)
```

 Build a Simple Hypergraph Transformer in PyTorch

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class HypKGTransformer(nn.Module):
    def __init__(self, embedding_dim=50, num_classes=25, num_heads=5, dropout=0.2):
        super().__init__()
        self.attn = nn.MultiheadAttention(embed_dim=embedding_dim, num_heads=num_heads, batch_first=True)
        self.dropout = nn.Dropout(dropout)
        self.norm = nn.LayerNorm(embedding_dim)
        self.fc = nn.Sequential(
            nn.Linear(embedding_dim, embedding_dim),
            nn.ReLU(),
            nn.Linear(embedding_dim, num_classes)
        )

    def forward(self, x, mask):
        # mask: 1 = real token, 0 = pad → attention mask needs 0 = keep, -inf = mask
        attn_mask = (~mask.bool()).unsqueeze(1).expand(-1, mask.shape[1], -1)
        attn_out, _ = self.attn(x, x, x, key_padding_mask=~mask.bool())
        x = self.norm(x + self.dropout(attn_out))
        x = x * mask.unsqueeze(-1)  # mask padded rows
        x = x.sum(dim=1) / mask.sum(dim=1, keepdim=True)  # mean pooling
        return self.fc(x)
```

```python
from torch.utils.data import TensorDataset, DataLoader

# Hyperparameters
batch_size = 16
lr = 1e-3
epochs = 200
embedding_dim = X.shape[-1]
num_classes = labels.shape[1]

# Dataloader
dataset = TensorDataset(X, M, labels)
loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Model, loss, optimizer
# Model, loss, optimizer
model = HypKGTransformer(embedding_dim=embedding_dim, num_classes=num_classes, num_heads=5)
criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
```

```python
for epoch in range(epochs):
    model.train()
    total_loss = 0

    for x_batch, m_batch, y_batch in loader:
        preds = model(x_batch, m_batch)
        loss = criterion(preds, y_batch.float())

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}/{epochs} - Loss: {total_loss/len(loader):.4f}")
```

Avaluate the Model

```python
from sklearn.metrics import f1_score, roc_auc_score
import numpy as np

# Put model in eval mode
model.eval()

# Collect predictions and true labels
all_preds = []
all_probs = []
all_labels = []

with torch.no_grad():
    for x_batch, m_batch, y_batch in loader:
        logits = model(x_batch, m_batch)
        probs = torch.sigmoid(logits)
        preds = (probs > 0.5).int()

        all_preds.append(preds.numpy())
        all_probs.append(probs.numpy())
        all_labels.append(y_batch.numpy())

# Concatenate all batches
y_true = np.vstack(all_labels)
y_pred = np.vstack(all_preds)
y_prob = np.vstack(all_probs)
```

```python
def precision_at_k(y_true, y_prob, k=5):
    """Compute mean Precision@k for a batch."""
    n_samples, n_classes = y_true.shape
    precision_scores = []

    for i in range(n_samples):
        top_k_indices = np.argsort(y_prob[i])[::-1][:k]
        relevant = y_true[i][top_k_indices]
        precision = relevant.sum() / k
        precision_scores.append(precision)

    return np.mean(precision_scores)
```

```python
from sklearn.metrics import f1_score, roc_auc_score

# Compute metrics
f1_micro = f1_score(y_true, y_pred, average="micro")
f1_macro = f1_score(y_true, y_pred, average="macro")

try:
    auc_micro = roc_auc_score(y_true, y_prob, average="micro")
    auc_macro = roc_auc_score(y_true, y_prob, average="macro")
except ValueError:
    auc_micro = auc_macro = "AUC undefined (insufficient class variance)"

p_at_3 = precision_at_k(y_true, y_prob, k=3)
p_at_5 = precision_at_k(y_true, y_prob, k=5)

print("Evaluation Results (HypKG-style):")
print(f"Micro F1 Score:  {f1_micro:.4f}")
print(f"Macro F1 Score:  {f1_macro:.4f}")
print(f"Micro AUC:       {auc_micro}")
print(f"Macro AUC:       {auc_macro}")
print(f"Precision@3:     {p_at_3:.4f}")
print(f"Precision@5:     {p_at_5:.4f}")
```
